```python
import numpy as np
```

```python
n = 26

dist_between_following_planets = np.array([
    100, 
    200,
    500,
    1000,
    2000,
    5000,
    10000,
    20000,
    50000,
    100000,
    200000,
    500000,
    1000000,
    2000000,
    5000000,
    10000000,
    20000000,
    50000000,
    100000000,
    200000000,
    500000000,
    1000000000,
    2000000000,
    5000000000,
    10000000000
], dtype=np.float64)  # n-1 distances
dists_matrix = np.zeros((n, n))
```

```python
for i in range(n-1):
    dists_matrix[i, i+1:] = np.sqrt(np.cumsum(dist_between_following_planets[i:]**2))

dists_matrix += dists_matrix.T
```
